```python
from scipy import optimize
import numpy as np
x = np.array([0,0.10,0.20,0.60])
y = np.array([1,0.50,0.20,0])
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
optimize.curve_fit(func,  x,  y)
```
